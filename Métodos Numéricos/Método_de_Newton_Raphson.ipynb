#<h1><font color="Orange">Código con Iteraciones</font></h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def fun(x):
   return x**3 + 2*x**2 + 10*x - 20

a = float(input("Digite el primer número: "))
max_iter = int(input("Digite la cantidad máxima de iteraciones: "))

# Definir la variable simbólica
x = sp.Symbol('x')

# Definir la derivada de la función
fun_deri = sp.diff(fun(x), x)

List_i = []
List_Xn = []
List_Error = []

Xn = a
i = 0
Error = 100

for _ in range(max_iter):
   List_i.append(i)
   List_Xn.append(Xn)

   Xn_new = Xn - (fun(Xn) / fun_deri.subs(x, Xn))

   if i >= 1:
       xnold = List_Xn[i-1]
       Error = np.abs((Xn_new - xnold) / Xn_new)
       List_Error.append(Error)
   else:
       Error = 100
       List_Error.append(Error)

   Xn = Xn_new
   i += 1

   print(f"En la iteración {i} Xn es {Xn.evalf(6)} con un error de {Error:.6f}")

x_vals = np.linspace(-5, 5, 100)
y_vals = [fun(x) for x in x_vals]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', label='$y=x^3 + 2x^2 + 10x-20$')
plt.plot(List_i, List_Xn, 'ro', label='Aproximaciones de la raíz')
plt.grid()
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función y aproximaciones de la raíz')
plt.show()
```

```python
plt.plot(List_i, List_Error, label = '$E_{n} vs i$')
plt.grid()
plt.legend(loc = 'right')
plt.show()
```

#<h1><font color="Orange">Código con Tolerancia</font></h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def fun(x):
    return x**3 + 2*x**2 + 10*x - 20

a = float(input("Digite el primer número: "))
tol = 1E-8

# Definir la variable simbólica
x = sp.Symbol('x')

# Definir la derivada de la función
fun_deri = sp.diff(fun(x), x)

List_i = []
List_Xn = []
List_Error = []

Xn = a
i = 0
Error = 100

while Error >= tol:
    List_i.append(i)
    List_Xn.append(Xn)

    Xn_new = Xn - (fun(Xn) / fun_deri.subs(x, Xn))

    if i >= 1:
        xnold = List_Xn[i-1]
        Error = np.abs((Xn_new - xnold) / Xn_new)
        List_Error.append(Error)
    else:
        Error = 100
        List_Error.append(Error)

    Xn = Xn_new
    i += 1

    print(f"En la iteración {i} Xn es {Xn:.6f} con un error de {Error:.6f}")

x_vals = np.linspace(-5, 5, 100)
y_vals = [fun(x) for x in x_vals]

x_vals = np.linspace(-5, 5, 100)
y_vals = [fun(x) for x in x_vals]

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, 'b-', label='$y=x^3 + 2x^2 + 10x-20$')
plt.plot(List_i, List_Xn, 'ro', label='Aproximaciones de la raíz')
plt.grid()
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función y aproximaciones de la raíz')
plt.show()
```

```python
plt.plot(List_i, List_Error, label = '$E_{n} vs i$')
plt.grid()
plt.legend(loc = 'right')
plt.show()
```
