#<h1><font color="Orange">Método de punto fijo</font></h1>

El método de punto fijo para encontrar una raíz α de la ecuación x = g(x), consiste en generar una sucesión {xn}n que se define mediante la fórmula de iteración: xn = g(xn-1), que converge a α cualquiera sea x0 ∈ [a, b]. La función g(x) se denomina función de iteración de punto fijo.

#<h1><font color="Orange">Código con Iteraciones</font></h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def g(x):
   return (- 2*(x**3) + 11.7*(x**2) + 5)/17.7 #Para la funcion original, se debe despejar x, es decir que debe quedar en terminos de x, ya sea sumando o pasando al otro lado la x

iter = int(input("Digite la cantidad de iteraciones: "))
lista_x = []
lista_i = []
lista_error = []
error = 100
x = 0

lista_error.append(error)

for i in range(iter):
  x = g(x)
  lista_x.append(x)
  lista_i.append(i)

  if i >= 1:
    xnold = lista_x[i-1]
    error = np.abs((x - xnold) / x)
    lista_error.append(error)

  print(f"En la iteración {i+1}, g(x) es {x}")

  if i >= 1:
    print(f"Error = {error}")

plt.plot(lista_i, lista_x, marker='o', color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.plot(lista_i, lista_error, label = '$E_{n} vs i$')
plt.grid()
plt.legend(loc = 'right')
plt.show()
```

#<h1><font color="Orange">Código con Tolerancia</font></h1>

```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return np.exp(-x)  # Ejemplo de función de iteración

tolerancia = 1E-8  # Tolerancia para la convergencia

# Variables para almacenar los valores de iteración
lista_x = []
lista_i = []
lista_error = []

error = 100
x = 0  # Valor inicial de x
i = 0  # Contador de iteraciones

lista_error.append(error)

while error >= tolerancia:
    x_new = g(x)  # Realizar la iteración utilizando la función de iteración g(x)
    lista_x.append(x_new)
    lista_i.append(i)

    if i >= 1:
        xnold = lista_x[i-1]
        error = np.abs((x_new - xnold) / x_new)
        lista_error.append(error)

    i += 1
    print(f"En la iteración {i}, x es {x_new}")

    x = x_new

    if i >= 2:
        print(f"Error = {error}")

# Graficar la convergencia de los valores de x
plt.plot(lista_i, lista_x, marker='o', color='b')
plt.title('Convergencia de xn')
plt.xlabel('Iteración')
plt.ylabel('xn')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
plt.plot(lista_i, lista_error, label = '$E_{n} vs i$')
plt.grid()
plt.legend(loc = 'right')
plt.show()
```
