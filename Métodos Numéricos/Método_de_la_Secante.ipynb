#<h1><font color="Orange">Método de la Secante</font></h1>
Es una variación del método de Newton-Raphson donde en vez de calcular la derivada de la función en el punto de estudio, teniendo en mente la definición de derivada, se aproxima la pendiente a la recta que une la función evaluada en el punto de estudio y en el punto de la iteración anterior.

#<h1><font color="Orange">Código con Iteraciones</font></h1>

```python
import numpy as np
import matplotlib.pyplot as plt

list_x = []
list_e = []
list_i = []
x = 5
iter = 3

def funcion(x):
    return -x**2 + 1.8*x + 2.5 #para -x^2 +1.8x + 2.5

for i in range(iter):
    x = funcion(x)
    list_x.append(x)
    list_i.append(i)

    if i >= 1:
        error = np.abs(((list_x[i] - list_x[i-1]) / list_x[i]))
    else:
        error = 100
    list_e.append(error)
    print("raiz: ", x, " error: ", error)

plt.figure(figsize=(8, 6))

x_vals = np.linspace(0, 4, 100)
y_vals = funcion(x_vals)
plt.title('Convergencia de xn')
plt.plot(x_vals, y_vals)

plt.grid(True)
plt.legend(loc='upper left')
plt.xlabel('iteraciones')
plt.ylabel('xn')

plt.show()
```

#<h1><font color="Orange">Código con Tolerancia</font></h1>


```python
import numpy as np
import matplotlib.pyplot as plt

def fun(x):
    #return x**3 + 2*x**2 + 10*x - 20
    return -x**2 + 1.8*x + 2.5
def m(a, b):
    return (fun(b) - fun(a)) / (b - a)

a = float(input("Digite el límite inferior del rango de búsqueda: "))
b = float(input("Digite el límite superior del rango de búsqueda: "))
tol = 1E-8

List_i = []
List_Xn = []
List_Error = []

i = 0
Error = 100

while Error >= tol:
    Xn = a - (fun(a) / m(a, b))

    List_i.append(i)
    List_Xn.append(Xn)

    if (Xn < a):
        b = a
        a = Xn
    else:
        a = b
        b = Xn

    if (i >= 1):
        xnold = List_Xn[i-1]
        Error = np.abs((Xn - xnold) / Xn)
    else:
        Error = 100
    List_Error.append(Error)

    print(f"En la iteración {i+1} Xn es {Xn:.6f} con un error de {Error:.6f}")
    i += 1

x_vals = np.linspace(min(a, b)-1, max(a, b)+1, 100)
y_vals = [fun(x) for x in x_vals]

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x_vals, y_vals)
ax.plot(List_Xn, [fun(xn) for xn in List_Xn], 'ro', label='Aproximaciones de la raíz')
ax.grid()
ax.legend(loc='upper left')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Gráfica de la función y aproximaciones de la raíz')
plt.show()
```

```python
plt.plot(List_i, List_Error, label = '$E_{n} vs i$')
plt.grid()
plt.legend(loc = 'right')
plt.show()
```
