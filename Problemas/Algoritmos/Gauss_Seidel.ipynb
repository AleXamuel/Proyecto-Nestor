```python
import numpy as np
import matplotlib.pyplot as plt

N = int(input("Digite las iteraciones: "))
M = np.array([[10, 2, -1, 0, 26], [1, 20, -2, 3, -15], [-2, 20, 30, 0, 53], [1, 2, 3, 20, 47]])
D = np.diagonal(M)
I = M[:, -1]
Id = np.eye(M.shape[0])
MD = M[:, 0:-1] - (D * Id)
P = np.array([1.0, 2.0, 3.0, 4.0])
ListP = []
ListError = []

for iter in range(N):
    for j in range(M.shape[0]):
        P[j] = (I[j] - np.dot(MD[j, :], P)) / D[j]
    ListP.append(P.copy())

    if iter != 0:
        Error = np.abs((ListP[iter] - ListP[iter-1])/ ListP[iter])
        ListError.append(list(Error))

    print(f"Iteración {iter+1}: {P}")
    print(f"Error: {Error}" if iter != 0 else "")
```

```python
for i in range(len(ListError[0])):
    plt.plot([row[i] for row in ListError], label=f'Error en P{i}')

plt.xlabel("Número de iteraciones")
plt.ylabel("Error")
plt.title("Evolución del error en función del número de iteraciones")
plt.show()
```
