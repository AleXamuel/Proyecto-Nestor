**binary exponentiation**

```python
def binpow(a,b,m):
    a%=m
    ans = 1
    while b>0:
        if b&1:
            ans*=a
            ans%=m
        a*=a
        a%=m
        b>>=1
    return ans
```

**Miller rabbin primality check**

```python
def test_base(a,n,d,s):
    cur = binpow(a,d,n)

    for _ in range(s):
        aux = (cur*cur)%n
        if aux==1:
            return cur==1 or cur==n-1
        cur = aux
    return cur==1

def miller_rabin(n):
    if n<=1:
        return False
    if (n&1)==0:
        return n==2

    d = n-1
    s = 0
    while (d&1)==0:
        s+=1
        d>>=1
    for a in [2, 3, 5, 7,11,13,17,23]:
        if n==a:
            return True
        if not test_base(a,n,d,s):
            return False
    return True
    
```

**Sieve**

```python
MAX = int(1e5)
primos = []
is_prime = [True]*MAX
is_prime[0] = is_prime[1] = False
for i in range(2,MAX):
    if is_prime[i]:
        primos.append(i)
        if i*i>=MAX:
            continue
        j = i*i
        while j<MAX:
            is_prime[j] = False
            j+=i
```

**Classic prime check**

```python
def is_prime(n):
    if n<=1:
        return False
    if n==2 or n==3 or n==5:
        return True
    if n%2==0 or n%3==0 or n%5==0:
        return False
    i = 5
    while i*i<=n:
        if n%i==0 or n%(i+2)==0:
            return False
        i+=6
    return True
```

**Factorize a number**

```python
def factorize(n):
    ans = []
    i = 2
    while i*i<=n:
        while n%i==0:
            ans.append(i)
            n//=i
        i+=1
    if n>1:
        ans.append(n)
    return ans
```

**Get divisors**

```python
def divs(n):
    ans = []
    i = 1
    while i*i<=n:
        if n%i==0:
            ans.append(i)
            if n//i!=i:
                ans.append(n//i)
        i+=1
    ans.sort()
    return ans
```
